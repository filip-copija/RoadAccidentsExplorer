```python
file_path = r"..\data\raw_accidents_2019_2024.csv"
```

```python
import pandas as pd

data = pd.read_csv(file_path, delimiter=";")
```

```python
missing_values = data.isnull().sum()
missing_values
```

```python
def determine_region_type(code):
    code = str(code).zfill(7)  
    if code == "0000000":  # Województwo 
        return "kraj"
    if code[2:] == "00000":  # Województwo
        return "województwo"
    elif code.isdigit() and len(code) == 7:  # Miasto
        return "miasto"
    else:
        return "nieznane"

data["Type"] = data["Kod"].apply(determine_region_type)
data
```

```python
regions = data[data["Type"].isin(["kraj", "województwo", "powiat"])]
unique_regions = regions[["Kod", "Nazwa", "Type"]].drop_duplicates()

unique_regions = unique_regions.sort_values(by=["Type", "Kod"])
df_sorted = unique_regions.sort_values(by="Kod", ascending=True)
df_sorted
```
